```python
#imported my libraries
# Data processing

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
df=sns.load_dataset("tips")
```

```python
# View the  5 Top rows
df.head()
```

```python
# View the bottom 5 rows
df.tail()
```

```python
# Inspect the DataSet
df.info()
```

```python
# Statistically Analysis of the DataSet
df.describe()
```

```python
#This graph displays Total Customer by Sex
plt.figure(figsize=(5,3))
sns.countplot(x=df['sex'])
```

```python
#This graph displays Total Customer by Smokers
plt.figure(figsize=(5,3))
sns.countplot(x=df['smoker'])
```

```python
#This graph displays Total Customer by Day
plt.figure(figsize=(5,3))
sns.countplot(y=df['day'])
```

```python
# Visualization of numerical column - Total Bills and Tips (boxplots)
fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(15,3))
sns.boxplot(df['total_bill'],ax=axes[0])
axes[0].set_title("TOTAL BILL")


sns.boxplot(df['tip'],ax=axes[1])
axes[1].set_title("TIPS")

```

```python
# Total Bill by Sex
total_BillSex = df.groupby('sex')['total_bill'].sum().reset_index()
sns.barplot(x='sex',y='total_bill',data=total_BillSex)
```
